# House Price Prediction
A house costs 50k + 50k per bedroom, so that a 1 bedroom house costs 100k, a 2 bedroom house costs 150k and so on.

Let's make a neural network that learns this relationship so that it would predict the pricing of the house with as many number of bedrooms as we want.


# Libraries

Importing all the libraries we would need.

```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

#Our Neural Network

```
def house_model(y_new):
    xs =np.array ([1,2,3,4,5,6], dtype=float)
    ys =np.array ([1,1.5,2,2.5,3,3.5], dtype=float)
    model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
    model.compile(optimizer='sgd', loss='mean_squared_error')
    model.fit(xs, ys, epochs=500)
    return model.predict(y_new)[0]
```

# Time for prediction

```
prediction = house_model([7.0])
print(prediction)
```
